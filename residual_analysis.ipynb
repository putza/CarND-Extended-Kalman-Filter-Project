```python
import pandas as pd
import scipy as sp
import pandas as pd

import pylab as plt
%matplotlib notebook
```

```python
df = pd.read_csv('build/residuals.csv',header=None,skiprows=3)
col_headers = ['px','py','vx','vy']
```

```python
df_gt = df[[0,1,2,3]]
df_kf = df[[4,5,6,7]]
df_e1 = df[[8,9,10,11]]

df_gt.columns = col_headers
df_kf.columns = col_headers
df_e1.columns = col_headers
```

```python
df_gt.plot()
df_kf.plot()
```

```python
plt.figure(figsize=(12,12))
plt.plot(df_gt['px'],df_gt['py'],'g-',linewidth=5)
plt.plot(df_kf['px'],df_kf['py'],'r.')
plt.quiver(df_kf['px'],df_kf['py'],df_kf['vx'],df_kf['vy'])
```

```python
# Calculate Error

res  = df_gt - df_kf
rmse = sp.sqrt((res**2).sum()/len(res))
```
